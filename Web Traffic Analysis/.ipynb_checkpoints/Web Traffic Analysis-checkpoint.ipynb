```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

# Introduction

The dataset was taken from kaggle at https://www.kaggle.com/datasets/bobnau/daily-website-visitors." The data set contains 5 years of daily time series data for several measures of traffic on a statistical forecasting teaching notes website whose alias is statforecasting.com". Below we could see a snippet of the data.

```python
data = pd.read_csv("daily-website-visitors.csv")
data.head()
```

### Overview of the Data

- Row: Refers to the row number
- Day: Day of the week in which the data was recorded
- Day.Of.Week: same as above but in numeric form where Sunday = 1, Monday = 2 and so on
- Date: Date in which the data was collected
- Page.Loads: Number of time the page was visited
- Unique.Visits: Number of unique users that visited the page (based on IP address which can undercount the real number of unique visits)
- First.Time.Visits: Number of users that visited the page for the first time (again based on IP address)
- Returning.Visits: Number of non-first time users, important to note that Returning.Visits = First.Time.Visits - Unique.Visits

Let's look at the type of each data to see if we have to do any data cleaning.

# Data Cleaning

```python
data.dtypes
```

As we can, the types for some of the columns are not correct so let's convert them below.

```python
#we need to change the type Date to datetime and Page.Loads, Unique.Visits, First.Time.Visits, and Returning.Visits 
# to int32 (could also change Day.Of.Week to int32 to save space)

data['Date'] = pd.to_datetime(data['Date'])
data['Day.Of.Week'] = data['Day.Of.Week'].astype('int32')

for col in ['Page.Loads', 'Unique.Visits', 'First.Time.Visits', 'Returning.Visits']:
    data[col] = data[col].map(lambda x: x.replace(',', '')).astype('int32')
    

data.dtypes
```

This looks much better. Let's look at some basic statistics to see if anything standouts (i.e missing values, outliers, inbalances, etc.)

```python
data.describe()

# looks fine to me
```

Below let's engineer some features that we think would provide some valuable insight towards the patterns in the data.

```python
#feature engineering

#non-unique visits
data['Non-unique visits'] = data['Page.Loads'] - data['Unique.Visits']

#percentage for each category
data['Percent Unique'] = data['Unique.Visits'] / data['Page.Loads']
data['Percent Non-unique'] = 1 - data['Percent Unique']
data['Percent First-Time'] = data['First.Time.Visits'] / data['Unique.Visits']
data['Percent Returning'] = 1 - data['Percent First-Time']

#is the date a weekend or not
data['Weekend?'] = data['Day.Of.Week'].map(lambda x: int((x == 7) or (x == 1)))

#week number
def weekno(x):
    return x.isocalendar()[1] + (x.year - 2014) * 52
data['weekno'] = data['Date'].map(lambda x: weekno(x))

#month number
def monthno(x):
    return x.month + (x.year - 2014) * 12

data['monthno'] = data['Date'].map(lambda x: monthno(x))
data.head()
```

Looks pretty solid, let's check for some basic statistics one last time before we start looking at some visualiztions

```python
data.describe()
```

Looks good. Now let's look at the distribution of the data to give us a better understanding of the patterns in the data.

Let's try to identify trends in the data now

```python
#weekend vs weekday trends

by_weekend = data.groupby('Weekend?')[['Page.Loads', 'Unique.Visits', 'First.Time.Visits', 'Returning.Visits', 'Non-unique visits']].sum()
by_weekend.head()
```

We expected the weekend to have less total interaction compared to weekday, since there are simply more weekdays than weekends so let's normalize it.


```python
by_weekend.iloc[0] = by_weekend.iloc[0].apply(lambda x: x/5)
by_weekend.iloc[1] = by_weekend.iloc[1].apply(lambda x: x/2)
by_weekend
```

Whoa, there's still quite a bit of a difference. For example, if we focus on page.loads, we could see that that's quite a (approximately 40% less for each column). The other columns follow in suit. We could run a basic permutation test to test the statistical significance of this.

Hypothesis: There are more Page.Loads on the weekdays than there is on weekends.

Null Hypothese: There are an equal amount of Page.Loads on the weekdays compared to the weekends.


```python
counts = 0

observed_prop = by_weekend.iloc[1]['Page.Loads']/by_weekend.iloc[0]['Page.Loads']

pageloads = data[['Weekend?', 'Page.Loads']]

for i in range(1000):
    pageloads['Page.Loads'] = pageloads['Page.Loads'].sample(frac = 1).values
    grouped = pageloads.groupby("Weekend?").sum()
    grouped.iloc[0] = grouped.iloc[0].apply(lambda x: x/5)
    grouped.iloc[1] = grouped.iloc[1].apply(lambda x: x/2)
    curr = grouped.iloc[1]['Page.Loads']/grouped.iloc[0]['Page.Loads']
    if (curr <= observed_prop):
        counts += 1
```

```python
print(counts / 1000)
```

We conclude that there is high statistical evidence that what we witnessed is not due to random chance -> weekend significantly less clicks than weekdays. If we reason through this logicially, this should make sense since people usually spend more time on work related activities during the weekday over the weekend. This applies to our particular website since it's a website on lecture notes.

# Prediction Task

As we said earlier, since Page.Loads is a pretty good indicator for most of the other variables in the dataset, we'll try to predict number of Page.Loads on a date in the future.

```python
# naive forecast model

data['predictions'] = data['Page.Loads'].shift(1)
```

```python
#plot predictions
sns.lineplot(data[['Date', 'Page.Loads', 'predictions']])
```

This is kind of hard to see, so we'll just focus on the last 100 days

```python
sns.lineplot(data[['Date', 'Page.Loads', 'predictions']][-100:])
```

Let's calculate the MAE for this so we have a numeric baseline to beat (we will only look at the last 100 days for our test set moving forward)


```python
1/100 * (abs(data['Page.Loads'][-100:] - data['predictions'][-100:])).sum()
```
